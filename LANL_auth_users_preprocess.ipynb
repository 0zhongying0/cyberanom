## Data processing phase 1 for LANL auth.txt file 

In this notebook we do few things

1. Analysis of the redteam data.
2. extract events for each user who are also have events in the redteam.txt file
3. extract events for the first 100 users who are not in the redream.txt file.

This should allow us to process smaller files further (see Data processing phase 2) for model input 


```python
auth_fname = '/Volumes/TheFarSide/Data/LANL_data/auth.txt'
users_dir = '/Volumes/TheFarSide/Data/LANL_data/users'
redteam_fname = '/Volumes/TheFarSide/Data/LANL_data/redteam.txt'
```

```python
# find out the number of users in the redteam.txt
users = {}

with open(redteam_fname, 'r') as red:
    for line in red:
        items = line.strip().split(',')
        user = items[1].strip().split('@')[0]
        if user in users:
            users[user] +=1
        else:
            users[user] = 1

print('num users in redteam.txt:', len(users.keys()))
#print(users)
        
```

```python
users['U748']
```

```python
# create files for all users in redteam.txt
users_files = {}
for k,v in users.items():
    fname = '{0}/{1}.txt'.format(users_dir, k)
    f = open(fname, 'w')
    users_files[k] = f
```

```python
# process auth.txt for all users's events in the redteam.txt and extract their data
count = 0

with open(auth_fname, 'r') as auth:
    for line in auth:
        items = line.strip().split(',')
        user = items[1].strip().split('@')[0]
        if user in users_files:
            users_files[user].write(line)
        count+=1
        if (count % 1000000)==0:
            print('processed: {0} lines'.format(count))

```

```python
for _,v in users_files.items():
    v.close()
```

```python
# process auth.txt for first 100 users's events not in the redteam.txt and extract their data
# the results will be in users_norm directory

users_norm_dir = '/Volumes/TheFarSide/Data/LANL_data/users_norm'
users_norm_files = {}

count = 0
num_users = 0

with open(auth_fname, 'r') as auth:
    for line in auth:
        items = line.strip().split(',')
        user = items[1].strip().split('@')[0]
        
        if user.startswith('U') and user not in users:
            if user in users_norm_files:
                users_norm_files[user].write(line)
            elif num_users < 100:
                fname = '{0}/{1}.txt'.format(users_norm_dir, user)
                f = open(fname, 'w')
                users_norm_files[user] = f
                f.write(line)
                num_users +=1
                
        count+=1
        if (count % 2000000)==0:
            print('processed: {0} lines'.format(count))

    for _,v in users_norm_files.items():
        v.close()
    print("num_users: ", num_users)
```

```python
all_users = set()
count = 0
with open(auth_fname, 'r') as auth:
    for line in auth:
        items = line.strip().split(',')
        user = items[1].strip().split('@')[0]
        if user.startswith('U'):
            all_users.add(user)
        count +=1 
        if (count % 10000000)==0:
            print('processed: {0} lines'.format(count))

print('all users in auth.txt: ', len(all_users))               

```
